```python
Name = input("Please input your name here ")
scoreattempts = int(input("How many quiz scores would you like to enter? "))

quiz_scores_list = []
attempts = 0
while attempts < scoreattempts:
    score = int(input("Please input your quiz scores "))
    quiz_scores_list.append(score)
    attempts += 1

total = sum(quiz_scores_list)
print("The total is", total)
print("The average score is ", total/scoreattempts)

if total/scoreattempts >=75:
  print("Passed")
elif total/scoreattempts <75:
  print("Failed")
```
